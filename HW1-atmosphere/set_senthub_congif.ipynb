# Create sentinelhub config file

```python
from sentinelhub import SHConfig
from dotenv import load_dotenv
import os
```

```python
load_dotenv()
```

```python
config = SHConfig()
config
```

```python
# Set instance ID, user ID and user SECRET
config.instance_id = 'HW1-instance'
config.sh_base_url = 'https://sh.dataspace.copernicus.eu'
config.sh_token_url = 'https://identity.dataspace.copernicus.eu/auth/realms/CDSE/protocol/openid-connect/token'
config.sh_client_id = os.environ.get("CDSE_CLIENT_ID")
config.sh_client_secret = os.environ.get("CDSE_CLIENT_SECRET")
config.save("myHW1-profile")
```
