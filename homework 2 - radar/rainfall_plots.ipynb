```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
radar_meta = np.load('data/radar_rainfall/rainfall_intensities/radar_metadata.npy')
radar_ranges = np.load('data/radar_rainfall/rainfall_intensities/ranges.npy')
radar_azimuths = np.load('data/radar_rainfall/rainfall_intensities/azimuths.npy')
```

```python
rainfall_202311130300 = np.load('data/radar_rainfall/accumulated_rainfall/1h/rainfall_202311130300.npy')
rainfall_202311130400 = np.load('data/radar_rainfall/accumulated_rainfall/1h/rainfall_202311130400.npy')
```

```python
r, theta = np.meshgrid(radar_ranges, np.radians(radar_azimuths))
fig, ax = plt.subplots( subplot_kw={'projection': 'polar'}, figsize=(8, 8))
c = ax.pcolormesh(theta, r, rainfall_202311130300 - rainfall_202311130400, shading='auto', cmap='viridis')  # Use any colormap you like
#ax.set_theta_zero_location("N")  # Zero degrees at the top
#ax.set_theta_direction(-1)       # Clockwise
#ax.set_title("Weather Radar Data (Polar View)", fontsize=14)
plt.colorbar(c, ax=ax, label='Reflectivity (?)')  # You can label it depending on your data

#c = ax[1].pcolormesh(theta, r, np.log(rainfall_202311130400), shading='auto', cmap='viridis')  # Use any colormap you like
#plt.colorbar(c, ax=ax[1], label='Reflectivity (?)')  # You can label it depending on your data
```
